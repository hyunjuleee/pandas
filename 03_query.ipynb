```python
tips = sns.load_dataset('tips')
tips.head()
```

### 1
- 남성 흡연자 조건 필터링
- tip을 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'

tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

### 2

```python
cond4 = tips['day'].isin(['Sat', 'Sun'])
```
