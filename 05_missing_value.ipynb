### fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'].fillna(100).tail()
```

```python
df1['age'] = df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
df1['deck']
```

```python
df1['deck'].fillna('C')
```

```python
df1['deck'].head()
```

```python
# df['deck'].fillna('No Data')
# 카테고리타입은 원 카테고리 범위에서 벗어나는 정보를 넣을 수 없음
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치 => 남자 승객의 평균으로 채우기
- 여자 승객의 결측치 => 여자 승객의 평균으로~

```python
df1 = df.copy()
```

```python
c1 = df1['sex'] == 'male'
c2 = df1['sex'] == 'female'
```

```python
m_mean = df1.loc[c1, 'age'].mean()
```

```python
df1.loc[c1, 'age'].fillna(m_mean)
df1.loc[c1, 'age'] = df1.loc[c1, 'age'].fillna(m_mean) # 할당
```

```python
f_mean = df1.loc[c2, 'age'].mean()
```

```python
df1.loc[c2, 'age'].fillna(f_mean)
df1.loc[c2, 'age'] = df1.loc[c2, 'age'].fillna(f_mean)
```
