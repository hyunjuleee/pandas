### to_numeric

```python
df1['운동량'] = pd.to_numeric(df1['운동량'], errors='coerce')

#숫자로 바꿀수 없는 데이터가 섞여 있어도 errors='coerce'로 나머지는 데이터는 숫자로 바꿀 수 있는 것
```

```python
df1.info()
```

```python
df1.tail()
```

```python
bins = [0, 1000, 2000, df1['운동량'].max()]
pd.cut(df1['운동량'], bins).value_counts()
```

```python
pd.cut(df1['운동량'], bins, labels=['부족', '보통', '많음'])
```

```python
df1.describe()
```

```python
df1['운동량_cut'] = pd.cut(df1['운동량'], bins=10)
df1['운동량_cut'].value_counts()
```

```python
df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10) # 비율 기준
```

```python
df1['운동량_qcut'].value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
```

```python
pd.qcut(df1['운동량'], qcut_bins)
```

```python
pd.qcut(df1['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
```
